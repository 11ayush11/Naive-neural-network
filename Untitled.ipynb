```python
from sklearn import linear_model
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math 
from scipy import stats as st 
```

```python
training_set = pd.read_csv('G:/ml/simple reg/train.csv')
test_set = pd.read_csv('G:/ml/simple reg/test.csv')
train_set = training_set.dropna()
test_set = test_set.dropna()

x_training_set = train_set.as_matrix(['x'])
y_training_set = train_set.as_matrix(['y'])

x_test_set = test_set.as_matrix(['x'])
y_test_set = test_set.as_matrix(['y'])

plt.title('Relationship between X and Y')
plt.scatter(x_training_set, y_training_set,  color='gray')
plt.show()
```

```python
model_linear = linear_model.LinearRegression()
model_linear.fit(x_training_set,y_training_set)
y_predicted = model_linear.predict(x_test_set)

print('R sq: ',model_linear.score(x_training_set,y_training_set))

# and so the correlation is..
print('Correlation: ', math.sqrt(model_linear.score(x_training_set,y_training_set)))

# Equation coefficient and Intercept
print("Coefficient for X: ", model_linear.coef_)

print("Intercept for X: ",  model_linear.intercept_)

plt.title('Comparison of Y values in test and the Predicted values')
plt.ylabel('Test Set')
plt.xlabel('Predicted values')
plt.plot(y_predicted, '.', y_test_set, 'x')
plt.show()
```
